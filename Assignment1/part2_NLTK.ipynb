```python
from nltk.book import *    #importing the content (texts) from the nltk book 
```

```python
from nltk import re       #importing regular expression from nltk
```

```python
md_tokens = text7.tokens        #extracting tokens from text7
md_tokens
```

```python
md_tokens_nopunct = [word.lower() for word in md_tokens if re.search("\w",word)]    #converting the tokens to lowercase to get the normalised tokens
md_tokens_nopunct     #printing tokens in lowercase
```

```python
from nltk import FreqDist
md_tokens_fdist = FreqDist(md_tokens_nopunct)      #checking frequency distribution for each token
md_tokens_fdist                                    #printing the frequency for each token
```

```python
md_tokens_fdist.most_common(50)            # finding and printing the most common 50 words
```
